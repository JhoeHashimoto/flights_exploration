```python
df_original = flights.copy()

cols = [
    'CANCELLATION_REASON',
    'AIR_SYSTEM_DELAY',
    'SECURITY_DELAY',
    'AIRLINE_DELAY',
    'LATE_AIRCRAFT_DELAY',
    'WEATHER_DELAY'
]

df_filtrado = df_original[df_original[cols].notna().any(axis=1)]

df_filtrado.head(10)
```

```python
################################
# TRATAMENTO DE VALORES AUSENTES
################################

df = flights.copy()

# Remover VOOS Cancelados
df = df[df["CANCELLED"] == 0].copy()

# Remover colunas com muitos nulos
threshold = 0.8
null_percent = df.isnull().mean()

print(null_percent)

cols_to_drop = null_percent[null_percent > threshold].index
df = df.drop(columns=cols_to_drop)

print("Colunas removidas:")
print(cols_to_drop.tolist())

# Preencher valores numéricos corretamente (SEM inplace)
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())


print("\nValores ausentes após tratamento:")
print(df.isnull().sum().sum())
```

```python
##########################
# CRIAÇÃO DA VARIÁVEL ALVO
##########################

df["ATRASO"] = (df["ARRIVAL_DELAY"] > 0).astype(int)

print(df["ATRASO"].value_counts())

#########################
# DISTRIBUIÇÃO DE ATRASOS
#########################

ax = sns.countplot(x="ATRASO", data=df)
df_curated = df.copy()
total = len(df)

for p in ax.patches:
    count = int(p.get_height())
    percentage = f"{100 * count / total:.1f}%"
    ax.annotate(percentage, 
                (p.get_x() + p.get_width() / 2., count),
                ha='center', va='bottom')

plt.title("Distribuição de Voos com Atraso")
plt.xlabel("Atraso (0 = Não, 1 = Sim)")
plt.ylabel("Quantidade")
plt.show()
```

## Inicio Aprendizado Supervisonado

```python
df_1 = df.copy()
df_2 = df.copy()
```

```python
df_1.shape[0]
```

```python
df_1.columns
```

```python
df_1.head()
```

```python
#####################
# SELEÇÃO DE FEATURES
#####################

cols_to_remove = [
    'ARRIVAL_DELAY',
    'ARRIVAL_TIME',
    'WHEELS_ON',
    'TAXI_IN',
    'SCHEDULED_ARRIVAL',
    'ELAPSED_TIME',
    'DEPARTURE_TIME',
    'YEAR',
    'DAY',
    'FLIGHT_NUMBER',
    'DIVERTED',
    'CANCELLED',
    'DEPARTURE_DELAY'
]


df_model = df_1.drop(columns=cols_to_remove, errors="ignore")

print("Colunas restantes:", df_model.shape[1])
df_print = df_model.columns
print(df_print)
######################
# AMOSTRAGEM DOS DADOS
######################

df_model = df_model.sample(n=100000, random_state=42)
print("Novo tamanho:", df_model.shape)
print("100K linhas")
# 100K Linhas 

########################
# SEPARAÇÃO DE VARIÁVEIS
########################

X = df_model.drop("ATRASO", axis=1)                  ################---->COLUNAS TREINAMENTOS
y = df_model["ATRASO"]                               ################--->TARGET


##########################
# REDUÇÃO DE CARDINALIDADE
###########################

#Trazer apenas 10 classificadores de cada campo (aplicar generalização para reduzir esforço de treinamento)
def reduzir_categorias(df, col, top_n=10):
    top = df[col].value_counts().nlargest(top_n).index
    df[col] = df[col].where(df[col].isin(top), "OUTROS")
    return df
    #TOP 10 categorias de cada campo 
cols_categoricas = X.select_dtypes(include="object").columns

for col in cols_categoricas:
    X = reduzir_categorias(X, col, top_n=10)

###################################
# ENCODING DE VARIÁVEIS CATEGÓRICAS (transformação de classes de string para número)
###################################

X = pd.get_dummies(X, drop_first=False)

#######################
# OTIMIZAÇÃO DE MEMÓRIA
#######################

X = X.astype("float32")

# remover colunas com pouca variância (remoção de categorias com baixa representatividade)
X = X.loc[:, X.mean() > 0.001]

print("Novo shape X:", X.shape)
X_2 = X.copy()
y_2 = y.copy()
```

```python
####################
# TRAIN / TEST SPLIT
####################

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,                                    # 80% treino e 20% teste
    random_state=42,
    stratify=y
)

print("Treino:", X_train.shape)
print("Teste:", X_test.shape)
```

```python
#### APLICAÇÃO MODELO REGRESSÃO

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

MODELO DE REGRESSÃO LOGÍSTICA – PREDIÇÃO DE ATRASOS

OBJETIVO:
Construir um modelo supervisionado para prever a probabilidade de atraso (ATRASO) com base em variáveis operacionais do voo, utilizando regressão logística.

PREPARAÇÃO DOS DADOS

° remoção de variáveis com vazamento de informação (ex: delays diretos e horários de chegada)
° seleção de variáveis estruturais: tempo, distância, horário, companhia e aeroportos
° amostragem de 100 mil registros para controle de performance
° redução de cardinalidade (top 10 categorias + “OUTROS”)
° aplicação de one-hot encoding para variáveis categóricas
° remoção de baixa variância
° dataset final com 42 features

DIVISÃO DOS DADOS

° treino: 80% (80.000 registros)
° teste: 20% (20.000 registros)
° estratificação pela variável alvo (ATRASO)

RESULTADOS DO MODELO

0 (sem atraso)	0.77	0.90	0.83
1 (com atraso)	0.75	0.54	0.63

° acurácia geral: 77%
° f1-score médio: 0.73

INTERPRETAÇÃO DOS RESULTADOS

° boa performance para classe 0 (voos sem atraso)
° modelo identifica corretamente a maioria dos voos pontuais (alto recall: 90%)

° performance moderada para classe 1 (voos com atraso)
° recall de 54% indica dificuldade em capturar todos os atrasos
° modelo tende a subestimar atrasos

PRINCIPAIS INSIGHTS

° o modelo é mais eficiente em identificar voos que NÃO atrasam
° existe perda de sensibilidade na detecção de atrasos (classe minoritária relativa)
° variáveis operacionais possuem poder preditivo relevante, mas não suficiente para capturar toda a variabilidade do atraso

LIMITAÇÕES IDENTIFICADAS

° ausência de normalização dos dados (impacta convergência do modelo)
° aviso de não convergência indica necessidade de ajuste de hiperparâmetros
° regressão logística pode não capturar relações não lineares complexas
° possível necessidade de balanceamento ou ajuste de threshold

CONCLUSÃO

O modelo apresentou desempenho consistente como baseline, com boa capacidade de identificar voos pontuais. No entanto, ainda possui limitação na detecção de atrasos, indicando oportunidade de melhoria com técnicas mais robustas ou ajustes no pré-processamento.

## Acurária e Precisão deram valores muito altos, vamos aplicar o modelo retirando algumas features

```python
##################
# REMOVER LEAKAGE 
##################

cols_leakage = [
    "TAXI_OUT",
    "AIR_TIME"
]

X_2 = X_2.drop(columns=cols_leakage)

####################
# TRAIN / TEST SPLIT
####################

from sklearn.model_selection import train_test_split

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(
    X_2, y_2, 
    test_size=0.2,
    random_state=42,
    stratify=y
)

###################
# FEATURE SELECTION 
###################

from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

feature_names = X_train_2.columns

selector = SelectKBest(score_func=f_classif, k=min(30, X_train_2.shape[1]))

X_train = selector.fit_transform(X_train_2, y_train_2)
X_test = selector.transform(X_test_2)

# análise
selected_mask = selector.get_support()

selected_features = feature_names[selected_mask]
removed_features = feature_names[~selected_mask]

df_features = pd.DataFrame({
    "feature": feature_names,
    "score": selector.scores_,
    "selecionada": selected_mask
}).sort_values(by="score", ascending=False)

print("📊 RESUMO:")
print(f"Selecionadas: {len(selected_features)}")
print(f"Removidas: {len(removed_features)}")
```

```python
#### APLICAÇÃO MODELO REGRESSÃO

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    RocCurveDisplay,
    precision_recall_curve,
    auc
)
import matplotlib.pyplot as plt

# predição
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

# -------------------------
# 1. Classification Report
# -------------------------
print("📊 CLASSIFICATION REPORT")
print(classification_report(y_test, y_pred))

# -------------------------
# 2. Matriz de confusão
# -------------------------
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Matriz de Confusão")
plt.show()

# -------------------------
# 3. ROC AUC
# -------------------------
roc_auc = roc_auc_score(y_test, y_prob)
print(f"\n🎯 ROC AUC: {roc_auc:.4f}")

RocCurveDisplay.from_predictions(y_test, y_prob)
plt.title("Curva ROC")
plt.show()

# -------------------------
# 4. Precision-Recall Curve
# -------------------------
precision, recall, _ = precision_recall_curve(y_test, y_prob)
pr_auc = auc(recall, precision)

print(f"📈 PR AUC: {pr_auc:.4f}")

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Curva Precision-Recall")
plt.show()
```

## Inicio Aprendizado Não Supervisionado

```python
##################
# CLUSTERING (KMEANS)
##################

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 1. Usar apenas X (sem target)
# -------------------------
X_unsup = X.copy()

# -------------------------
# 2. Escalar os dados
# -------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_unsup)

# -------------------------
# 3. Encontrar melhor número de clusters (Elbow)
# -------------------------
inertia = []

for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 10), inertia, marker='o')
plt.title("Método do Cotovelo (Elbow)")
plt.xlabel("Número de clusters")
plt.ylabel("Inertia")
plt.show()

# -------------------------
# 4. Treinar modelo final (ajuste k depois do gráfico)
# -------------------------
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# adicionar cluster ao dataset
df_clusters = pd.DataFrame(X_unsup.copy())
df_clusters["cluster"] = clusters

# -------------------------
# 5. Redução de dimensionalidade (PCA)
# -------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# -------------------------
# 6. Visualização
# -------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette="tab10")
plt.title("Clusters de Voos (PCA)")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.show()
```

### Clusterização fraca acima, Vamos agupar por cluster fazendo a média de cada um para identificar a probabilidade de atraso
